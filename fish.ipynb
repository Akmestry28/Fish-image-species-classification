```python
import tensorflow as tf  # For TensorFlow operations
import matplotlib.pyplot as plt  # For plotting images

train_dir = r'C:\Users\AKSHAD\Downloads\Dataset\images.cv_jzk6llhf18tm3k0kyttxz\data\train'
test_dir = r'C:\Users\AKSHAD\Downloads\Dataset\images.cv_jzk6llhf18tm3k0kyttxz\data\test'

img_size = (128,128)
batch_size = 32

train_ds = tf.keras.preprocessing.image_dataset_from_directory(
    train_dir,
    image_size = img_size,
    batch_size = batch_size
)

test_ds = tf.keras.preprocessing.image_dataset_from_directory(
    test_dir,
    image_size = img_size,
    batch_size = batch_size
)

print('lasses (Fish Species):',
      train_ds.class_names)

def show_images(dataset):
    plt.figure(figsize=(10,10))
    for images, labels in dataset.take(1):
        for i in range(9):
            plt.subplot(3, 3, i+1)
            plt.imshow(images[i].numpy().astype('uint8'))
            plt.title(dataset.class_names[labels[i]])
            plt.axis('off')
    plt.show()

show_images(train_ds)
show_images(test_ds)
```
